```python

import os
from nq_agents import data_extraction
from nq_agents import MultiAgentAutogen

my_agents = MultiAgentAutogen(
    llm_provider="ollama",
)

# read input
GOLD_FILE = 'data/v1.0-simplified_nq-dev-all_sample10_seed44.jsonl'
num_examples = data_extraction.count_lines(GOLD_FILE)
input_examples = data_extraction.read_first_lines(GOLD_FILE, num_examples)
print(f"Loaded {num_examples} examples from {GOLD_FILE}")

# predict
save_path = GOLD_FILE.replace('.jsonl', '_predictions.jsonl')
predictions = my_agents.predict_batch(
    examples=input_examples,
    save_path=save_path,
    verbose=True
)
```
